## Conceptes bàsics de Python
Amb Python podem definir els següents tipus de dades:



```python
# Nombres enters (o 'int')
my_int = 42

# Nombres decimals o de punt flotant (o 'float')
my_float = 3.14

# Cadenes de caràcters (o 'String')
my_string = "towel"

# Booleans (Dos valors possibles: True o False)
my_bool = True
```

Les tres estructures predefinides de Python són __listes__, __tuples__ i __diccionaris__. Aquestes estructures poden contenir elements de qualsevol tipus de dades

```python
my_list = [1, 2, "watermelon", ["whatever"]]     # Les llistes es defineixen usant []
my_tuple = (1, 2, 3, "watermelon", ["whatever"]) # Les tuples es defineixen usant ()
my_dict = {"key1": 1, "key2": [1,2,3]}           # Els diccionaris es defineixen usant {}
```

Es pot accedir als elements de les tres estructures anteriors usant []. Per a llistes i tuples, es necessari accedir usant l'índex de la posició. En canvi, per accedir a un diccionari, necessitem saber-ne la clau.

```python
print(my_list[2])        # Retorna un l'element de la llista situat a la posició 2 (tercer element de la llista ja que la primera posició té índex 0)
print(my_string[-2])     # També podem accedir des del final de la llista usant nombres negatius
print(my_tuple[1:3])     # O retornar un subconjunt, per exemple els elements amb índex 1 i 2. Observeu que l'últim element (índex 3) no es retorna
print(my_list[:-1])      # També podem retornar tots els elements fins a una certa posició. Per exemple tots excepte l'últim.
print(my_string[::2])    # Podem definir el 'pas' de la iteració. En aquest cas agafem els elements amb índex senar (1, 3 i 5)
print(my_string[::-1])   # Si el 'pas' és negatiu, els elements s'agafen en ordre invers.
print(my_dict['key2'])   # Per accedir als diccionaris és necessari conèixer-ne la clau
```

És important recordar que els objectes i estructures de Python es relacion mitjançant la seva referència, és a dir, no estem fent una còpia diferent de l'element. Així doncs, si dues variables apunten a un mateix element i el modifiquem en una d'elles, l'altre també es veurà afectada

```python
a = [1, 2, 3]
llista = ['A', 'B', 'C', 1, 2, 3, a]
print('Abans de la modificació:', llista)

a = a.append(4)
print('Després de la modificació:', llista)
```

És important que estigueu familiaritzats/des amb les __list comprehensions__ i __dict comprehensions__.
Són instruccions que, en una sola línia, permeten definir estructures de tipus __for__ de un o més nivells.

```python
a = [x for x in range(1,11)]                  # Nombres de l'1 al 10
b = [x for x in a if not x % 2]               # Nombres parells
c = [x**2 for x in b]                         # Potències quadràtiques dels elements parells

d = {x: [x-1, x+1] for x in a}                # Diccionari on cada clau té com a valor una llista amb el seu valor anterior i següent
e = {x: [i for i in range(x)] for x in a}     # Diccionari on cada clau conté com a valor una llista amb els elements >=0 anteriors a ell mateix (observa la doble iteracio for)
print(a)
print(b)
print(c)
print(d)
print(e)
```

<h1>Problema 1</h1>
<p><strong>
    Implementeu una funció, usant list comprehensions, que donada una variable entera <i>limit</i> i una altra variable entera <i>mult</i>, retorni una llista de tots els múltiples positius de <i>mult</i> menors o iguals a <i>limit</i>.
<br><br>Exemples:<br>
<i>mult_limit(10, 2) -> [2,4,6,8,10] <br>
mult_limit(15,3) -> [3, 6, 9, 12, 15]<br>
mult_limit(20,4) -> [4, 8, 12, 16, 20]<br>
mult_limit(5,3) -> [3]</i>
</strong></p>
</div>

```python
def mult_limit(limit,mult):
    # El teu codi aquí
    out_list = [x for x in range(1, limit + 1) if x%mult == 0]
    return out_list
```

```python
print(mult_limit(10,2))
print(mult_limit(15,3))
print(mult_limit(20,4))
print(mult_limit(5,3))
```

<h1>Problema 2</h1>
<p><strong>
Implementa una funció amb la mateixa sortida del problema anterior però que no retorni els mútliples d'un nombre enter <i>exc</i> passat com a paràmetre.
<br><br>
Ejemplos:<br>
<i>mult_limit_exc(10, 2, 4) -> [2,6,10]<br>
mult_limit_exc(20,3,9) -> [3, 6, 12, 15]<br>
mult_limit_exc(28,4,3) -> [4, 8, 16, 20, 28]<br>
mult_limit_exc(14,2,3) -> [2, 4, 8, 10, 14]</i>
</strong></p>
</div>

```python
def mult_limit_exc(limit, mult, exc):
    # El teu codi aquí
    out_list = [x for x in range(1, limit + 1) if x%mult == 0 and x%exc != 0]
    return out_list
```

```python
print(mult_limit_exc(10,2,4))
print(mult_limit_exc(20,3,9))
print(mult_limit_exc(28,4,3))
print(mult_limit_exc(14,2,3))
```

## Introducció als grafs

Un __graf__ es un conjunt definit com $G(V, E)$ on $V$ és un conjunt d'elements als que anomenem __vèrtexs__ o __nodes__ i $E$ és un conjunt de parelles d'elements de $V$ que anomenem __arestes__.

Podem representar un graf de forma gràfica representant els elements de $V$ en punts de l'espai i connectant-los seguint les instruccions de les arestes $E$. Per exemple, un graf on els conjunts $V$ i $E$ estan definits com:

$V = [1,2,3,4,5,6]$ <br>
$E = [(1,2), (1,5), (2,3), (2,5), (3,4), (4,5), (4,6)$

Pot ser representat pel diagrama següent:
![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/8/86/6n-graf-clique.svg/220px-6n-graf-clique.svg.png "Graph")

En un graf hi podem trobar molta informació (tant en el node com en l'aresta!). Per exemple, en un mapa de carreteres on els nodes son ciutats, cada node podria tenir les coordenades de la ciutat, la seva població, el país... I cada aresta podia contenir el nombre de benzineres, els quilòmetres, el tipus de carretera...

### Grafs amb pesos

Un dels tipus de grafs més utilitzats són els __grafs amb pesos__. Definim un graf amb pesos com a $G(V, E, P)$ on $P$ es un diccionari que té com a clau les arestes, $E$ i com a valors els pesos associats a cada aresta. Sovint representem els pesos d'un graf sobre l'aresta corresponent, com es pot veure a la figura següent.

![alt text](https://upload.wikimedia.org/wikipedia/commons/9/9a/Weighted_network.png "Graph")

### Grafs dirigits

Observa com en el graf del primer apartat, no distingim entre l'aresta $(3,4)$ i l'aresta $(4,3)$ ja que només tenim una sola aresta entre dos nodes i aquestes no tenen direcció especificada, simplement connecten els elements $3$ i $4$.

Això no sempre serà així ja que sovint volem representar relacions d'implicació en el graf o camins de sentit únic. Observa com ara l'aresta $(3,4)$ del següent graf és diferent a l'aresta $(4,3)$:

## NetworkX

[NetworkX](https://networkx.org/) és una llibreria molt útil per a generar i representar grafs. Té moltes funcions per estudiar la composició del graf i permet mostrar-los sense gaire esforç. A continuació fem una petita introducció en aquesta gran llibreria per conèixer-ne les funcions més bàsiques.

```python
import networkx as nx
```

### Creació d'un graf, nodes i arestes

```python
# Creació d'un graf sense cap element
G = nx.Graph()

# Inserim un node o nodes d'una llista
G.add_node(1)
G.add_nodes_from([2,3])

# Inserim arestes
# Observeu que estem afegint una aresta a un node que no existeix, el 4.
# Automàticament ens l'afegirà com a nou node
G.add_edges_from([(1,2), (2,3), (3,4)])

# Mostrem els nodes i les arestes del graf
print('Nodes de G:', G.nodes)
print('Arestes de G', G.edges)
```

```python
# Els nodes no han de ser necessariament nombres, poden ser altres tipus de variables
G = nx.Graph()
G.add_edges_from([('A',1), (1, 'B'), ('B', 'C'), (1, 'C')])

# Podem veure tots els nodes veïns a un node utilitzant la seva llista d'adjacència o
# cridant a la funcio neighbors()
print('Veïns del node 1:', list(G.adj[1]))
print('Veïns del node B:', list(G.neighbors('B')))
```

### Grafs dirigits, grafs amb pesos

```python
# Tambe podem crear grafs dirigits, usant:
DG = nx.DiGraph() # DiGraph => Directed Graph

# En qualsevol graf, tambe hi podem posar informació extra, a les arestes.
# En aquest cas, totes les arestes tindran els mateixos atributs
DG.add_edges_from([('A', 1), (1, 'B'), ('A', 'B')], mytext='soc_una_aresta', weight=0.5)

# Podem afegir arestes amb pesos usant una funció especial. Això ens permet posar pesos específics.
DG.add_weighted_edges_from([(1,2,0.1), (2,3,0.8), (1,3,0.4)])

# Per a retornar als veïns podem cridar la funcio neighbors() com hem fet anteriorment tot i que es
# recomana fer sevir la funció successors() que es específica de grafs dirigits.
# Observa el resultat següent. L'entens?
print('Successors (veïns) del node 1:', list(DG.successors(1)))

# Les arestes que acabem d'afegir, són arestes dirigides!
# Prova de descomentar i executar les següents línies una per una
print(G.edges['A',1])
print(G.edges[1,'A'])
print(DG.edges['A',1])
#print(DG.edges[1,'A'])
```

### Visualització

Podeu trobar exemples a "[Drawing](https://networkx.org/documentation/stable/auto_examples/index.html#drawing)".<br>
En el cas que obtingueu un error durant la visualització, executeu la comanda:

`pip3 install decorator -U`

i reinicieu el kernel

```python
import matplotlib.pyplot as plt

# Per a mostrar un graf, podem cridar la funcio draw()
# En aquest cas, li passem com a parametre extra que mostri el nom dels nodes i que ho posi en uns axis d'una figura
fig, axs = plt.subplots(1,2,figsize=(12,5))
nx.draw(G, with_labels=True, ax=axs[0])
axs[0].set_title('Graf simple')

# Per representar mes informació, necessitem extreure el paràmetre que ens interessi de l'aresta
labels = nx.get_edge_attributes(DG, 'weight')

# Com que haurem de pintar dos cops (un pel graf i l'altre per la informació de les arestes), hem de definir una posició ja que
# altrament, cada cop que es pinta el graf, generaria una estructura diferent.
pos=nx.spring_layout(DG)

# Mostrem el graf i, seguidament mostrem els pesos de les arestes
nx.draw(DG, pos=pos, with_labels=True)
nx.draw_networkx_edge_labels(DG, pos, edge_labels=labels, ax=axs[1])
axs[1].set_title('Graf dirigit amb pesos')

plt.show()
```

### Generació de grafs aleatoris

Una bona forma de provar que els vostres codis funcionen i són robustos al canviar de graf, és generar-ne d'aleatoris.
Podeu veure'n uns exemples a continuació. Trobareu la llista completa a "[Graph generators](https://networkx.org/documentation/stable/reference/generators.html)".

```python
# El mètode 'subplots()' permet generar una sola figura amb tants axis com li especifiquem en els dos primers paràmetres.
# En el nostre cas usem nrows=1, ncols=5. El paràmetre figsize permet definir la mida de la figura total resultant.
# Podem generar subplots amb més d'una fila però hem de tenir en compte que aleshores el paràmetre 'axs' que ens retorna
# serà una matriu i haurem d'accedir-hi amb dos índexs: axs[i][j]
fig, axs = plt.subplots(1,5,figsize=(30,5))

nx.draw(nx.binomial_graph(20,0.3), ax=axs[0])
axs[0].set_title('Binomial')

nx.draw(nx.newman_watts_strogatz_graph(20,3,0.2), ax=axs[1])
axs[1].set_title('Watts-Strogatz')

nx.draw(nx.powerlaw_cluster_graph(20,15,0.2), ax=axs[2])
axs[2].set_title('Holme and Kim powerlaw')

nx.draw(nx.random_lobster(20,0.7,0.6), ax=axs[3])
axs[3].set_title('Lobster')

nx.draw(nx.barabasi_albert_graph(20,3), ax=axs[4])
axs[4].set_title('Barabasi-Albert')

plt.show()
```
